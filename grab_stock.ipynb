```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
grab_df = pd.read_csv('e:/learning/data set/GRAB.csv')

grab_df.head()
```

```python
grab_df.shape
```

```python
grab_df.describe()
```

```python
fig = px.line(grab_df, x='Date', y=['Open', 'Close'])
fig.update_layout(title = dict( text = "<b>Open and Close Stock</b>", font = dict(size= 20, color='blue'), x = 0.5))
fig.update_yaxes(title_text='Open and Close')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=grab_df['Date'], y=grab_df['High'], mode='lines', name='High'))
fig.add_trace(go.Scatter(x=grab_df['Date'], y=grab_df['Low'], mode='lines', name="Low"))

fig.update_xaxes(title="Date")
fig.update_yaxes(title='Amount')

fig.update_layout(title = dict(text='<b>High and Low</b>', font= dict(size= 25, color='blue'), x=0.5))
fig.show()
```

```python
fig = px.histogram(grab_df, x= grab_df['Date'], y=grab_df['Volume'])

fig.show()
```

```python
grab_df['Date'] = pd.to_datetime(grab_df['Date'])
```

```python
grab_df['month_year'] = grab_df['Date'].dt.strftime('%Y-%m')
volume_by_month_year = grab_df.groupby('month_year')['Volume'].sum().reset_index(name='total')
```

```python
volume_by_month_year = volume_by_month_year.sort_values('month_year')
fig = px.bar(volume_by_month_year, x='month_year', y='total', color='month_year', text='total')

fig.update_xaxes(title='<b>Month of Year</b>')
fig.update_yaxes(title='<b>Volumes</b>')

fig.update_layout(title=dict(text='<b>Total Stock Sale by Month</b>',
                             font=dict(size=20, color='slateblue'), x=0.5),
                  xaxis=dict(tickfont=dict(color='blue')))
fig.show()
```
